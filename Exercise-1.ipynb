In this problem we want to determine whether the given mappings are vector norms. In order to determine this, we need to verify if they satisfy the three properties of a norm:

1. **Non-negativity**: $\|x\| \geq 0$ for all $x \in \mathbb{R}^d$ and $\|x\| = 0$ if and only if $x = 0$.
2. **Homogeneity**: $\|\alpha x\| = |\alpha| \|x\|$ for all $x \in \mathbb{R}^d$ and all scalars $\alpha \in \mathbb{R}$.
3. **Triangle inequality**: $\|x + y\| \leq \|x\| + \|y\|$ for all $x, y \in \mathbb{R}^d$.

### a) $\|x\|_{\frac{1}{2}} := \left( \sum_{k=1}^d |x_k|^{\frac{1}{2}} \right)^2$

1. **Non-negativity**: The expression $\sum_{k=1}^d |x_k|^{\frac{1}{2}}$ is non-negative because it is a sum of non-negative terms (absolute values raised to a positive power). Therefore, $\|x\|_{\frac{1}{2}} \geq 0$. However, $\|x\|_{\frac{1}{2}} = 0$ if and only if $x_k = 0$ for all $k$, which means for $x = 0$.

2. **Homogeneity**: For any scalar $\alpha \in \mathbb{R}$,

   $
   \|\alpha x\|_{\frac{1}{2}} = \left( \sum_{k=1}^d |\alpha x_k|^{\frac{1}{2}} \right)^2 = \left( |\alpha|^{\frac{1}{2}} \sum_{k=1}^d |x_k|^{\frac{1}{2}} \right)^2 = |\alpha| \left( \sum_{k=1}^d |x_k|^{\frac{1}{2}} \right)^2 = |\alpha| \|x\|_{\frac{1}{2}}.
   $

   The homogeneity property is satisfied.

3. **Triangle inequality**: To check the triangle inequality for $\|x\|_{\frac{1}{2}}$, consider two vectors $x, y \in \mathbb{R}^d$:

   $
   \|x + y\|_{\frac{1}{2}} = \left( \sum_{k=1}^d |x_k + y_k|^{\frac{1}{2}} \right)^2.
   $

   The triangle equality likely fails. We can show by using a counterexample:
   
   Let $x = (1, 0)$ and $y = (0, 1)$ be our vectors.

```python
def norm_half(x):
    return (sum(abs(xi)**0.5 for xi in x))**2

# Test the triangle inequality with x = (1, 0) and y = (0, 1)
x = [1, 0]
y = [0, 1]

# Compute norms
norm_x = norm_half(x)
norm_y = norm_half(y)
norm_x_plus_y = norm_half([x[i] + y[i] for i in range(len(x))])

# Print computed norms
print(f"||x||_(1/2): {norm_x}")
print(f"||y||_(1/2): {norm_y}")
print(f"||x + y||_(1/2): {norm_x_plus_y}")

# Check and print whether the triangle inequality holds
if norm_x_plus_y <= norm_x + norm_y:
    print("The triangle inequality holds.")
else:
    print("The triangle inequality does not hold.")
    print(f"{norm_x_plus_y} is not less than or equal to {norm_x + norm_y}")
```

### b) $\|x\|_0 := \#\{k : x_k \neq 0\}$ (the number of nonzero components of $x$)

1. **Non-negativity**:

    The function $\|x\|_0$ counts the number of non-zero components in the vector $x$.
    By definition, $\|x\|_0 \geq 0$ since the count of non-zero components cannot be negative.
    $\|x\|_0 = 0$ if and only if all components of $x$ are zero, i.e., $x = 0$. Thus, $\|x\|_0 = 0$ if and only if $x = 0$.
    Therefore, $\|x\|_0$ satisfies the non-negativity property.

2. **Homogeneity**:

    Multiplying a vector by a non-zero scalar $\alpha$ does not change the number of nonzero components, but multiplying the number of nonzero components by a non-zero scalar $\alpha$ does; so $\| \alpha \mathbf{x} \|_0 \neq |\alpha| \| \mathbf{x} \|_0$ unless $\alpha = 0$ or $|\alpha| = 1$. In order words, the homogeneity property is not satisfied.

Since $\|x\|_0$ does not satisfy all three properties it is **not** a norm.


### c) $\| \mathbf{x} \|_A := \| A\mathbf{x} \|_p$, where $A \in \mathbb{R}^{d \times d}$ and $p \in [1, \infty]$.

1. **Non-negativity:**   
   The $p$-norm $\| A\mathbf{x} \|_p$ is always non-negative, as it represents the sum of absolute values raised to a power $p \geq 1$.  
   Also, $\| \mathbf{x} \|_A = 0$ if and only if $A\mathbf{x} = \mathbf{0}$.  
   For this to imply $\mathbf{x} = \mathbf{0}$, the matrix $A$ must be invertible. So, $\| \mathbf{x} \|_A = 0$ if and only if $\mathbf{x} = \mathbf{0}$ when assuming $A$ is invertible.

2. **Homogeneity:**  
   For all $\mathbf{x} \in \mathbb{R}^d$ and all scalars $\alpha \in \mathbb{R}$,
   $
   \| \alpha \mathbf{x} \|_A = \| A(\alpha \mathbf{x}) \|_p = \| \alpha (A\mathbf{x}) \|_p = |\alpha| \| A\mathbf{x} \|_p = |\alpha| \| \mathbf{x} \|_A.
   $
   This shows that the homogeneity property is satisfied for all $\alpha \in \mathbb{R}$.


3. **Triangle inequality:**  
   For all $\mathbf{x}, \mathbf{y} \in \mathbb{R}^d$, 
   $
   \| \mathbf{x} + \mathbf{y} \|_A = \| A(\mathbf{x} + \mathbf{y}) \|_p \leq \| A\mathbf{x} \|_p + \| A\mathbf{y} \|_p = \| \mathbf{x} \|_A + \| \mathbf{y} \|_A.
   $
   This inequality holds because the $p$-norm $\| \cdot \|_p$ satisfies the triangle inequality. Therefore, the triangle inequality is as shown also satisfied for $\| \mathbf{x} \|_A$.

Since $\| \mathbf{x} \|_A = \| A\mathbf{x} \|_p$ satisfies all three properties it is a valid vector norm under the condition that $A$ is invertible.


### d) 

We want to show that, for $A \in \mathbb{R}^{m \times n}$, the Frobenius norm given by 

$$
\|A\|_F := \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij}|^2 \right)^{1/2}
$$

is a matrix norm.

1. **Non-negativity:**

   The norm must be non-negative, and it should be zero if and only if the matrix $A$ is the zero matrix.

    $$
    \| A \|_F = \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij}|^2 \right)^{1/2} \geq 0
    $$
    Since each $|a_{ij}|^2 \geq 0$, we know their sum is also non-negative. The square root of a non-negative number is non-negative.

  
    $$
    \| A \|_F = 0 \iff a_{ij} = 0 \quad \text{for all} \quad i, j
    $$
    The norm equals zero if and only if all elements $a_{ij}$ are zero, meaning that this condition holds if and only if $A$ is the zero matrix.

2. **Homogeneity:**

   Now we want to show homogeneity, meaning we want to show that for any scalar $\alpha \in \mathbb{R}$ and matrix $A \in \mathbb{R}^{m \times n}$, we have:

   $$
   \| \alpha A \|_F = |\alpha| \| A \|_F
   $$

     $$
     \| \alpha A \|_F = \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |\alpha a_{ij}|^2 \right)^{1/2} = \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |\alpha|^2 |a_{ij}|^2 \right)^{1/2}
     $$

     $$
     = \left( |\alpha|^2 \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij}|^2 \right)^{1/2} = |\alpha| \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij}|^2 \right)^{1/2} = |\alpha| \| A \|_F
     $$

3. **Triangle inequality:**

   The norm must satisfy the triangle inequality, meaning for any matrices $A, B \in \mathbb{R}^{m \times n}$, we have:

   $$
   \| A + B \|_F \leq \| A \|_F + \| B \|_F
   $$

   Using the definition of the Frobenius norm, we have:

   $$
   \| A + B \|_F = \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij} + b_{ij}|^2 \right)^{1/2}
   $$

   By the Minkowski inequality we have:

   $$
   \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij} + b_{ij}|^2 \right)^{1/2} \leq \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij}|^2 \right)^{1/2} + \left( \sum_{i=1}^{m} \sum_{j=1}^{n} |b_{ij}|^2 \right)^{1/2}
   $$

   So:

   $$
   \| A + B \|_F \leq \| A \|_F + \| B \|_F
   $$

Since the Frobenius norm $\| A \|_F$ satisfies the non-negativity, homogeneity, and triangle inequality properties, it is indeed a valid matrix norm.

To show that $(A,B) = trace(AB^T)$ defines an inner product on the vector space of $2x2$ real matrices we need to verify that it satisfies the four properties of an inner product.

*1) $\langle \lambda A + C, B \rangle = \lambda \langle A,C\rangle + \langle B,C\rangle$(linearity)*

* Trace is the sum of the diagonal elements of a matrix. Since 
\begin{equation}
trace(X + Y) = trace(X) + trace (Y), trace(\lambda X) = \lambda trace(X)
\end{equation}
for every $X, Y \in \mathbb{R^{nxn}} and \lambda \in \mathbb{R}.$ We have
\begin{equation} 
\langle \lambda A + C, B \rangle = trace(C^T (\lambda A + B)) = trace(\lambda C^T  A + C^T B) = \lambda trace(C^T A) + trace(C^T B) = \lambda \langle A,C\rangle + \langle B,C\rangle
\end{equation}


*2) $\langle A,B \rangle = \langle B,A \rangle$ (symmetry)*

* Since 
\begin{equation}
trace(X^T) = trace(X) 
\end{equation}
for every $X, Y \in \mathbb{R^{nxn}} and \lambda \in \mathbb{R}.$ We have
\begin{equation}
\langle A,B \rangle = trace(B^T A) = trace((B^T A)^T) = trace(A^T B) = \langle B,A \rangle
\end{equation}


*3) and 4) $\langle A,A \rangle > 0$ unless $A=0$ (positivity and definiteness)*
* For every $A=(A_{ij})$ $\in \mathbb{R^{mxn}}$ we have
\begin{equation} 
\langle A, A \rangle = trace(A^TA) = \sum_{i=1}^{n} (A^T A)_{ij} = \sum{i=1}^{n}\sum_{j=1}^{n} A^T_{ij} A_{ji} = \sum{i=1}^{n}\sum_{j=1}^{n} A^2_{ij} \geq 0
\end{equation}
and
\begin{equation} 
\langle A, A \rangle = \sum{i=1}^{n}\sum_{j=1}^{n} A^2_{ij} = 0 \Leftrightarrow (A_{ij} = 0  \forall  i,j) \Leftrightarrow A = 0
\end{equation}

All four properties of an inner product are satisfied, and $ \langle A, B \rangle = \text{trace}(A B^\top) $ defines an inner product on the vector space of $( 2 \times 2 )$ real matrices.

A symmetric positive-definite matrix, SPD, is a symmetric matrix $A \in \mathbb{R}^{nxn} $ such that $\bold{x}^T A\bold{x} > 0$ for all $\bold{x} \in \mathbb{R}^n \backslash \{0\}$. To show that the products $ABA$ and $BAB$ of SPD matrices $A,B \in \mathbb{R^{nxn}}$ are SPD matrices, we need to show that they are symmetric and positive-definite.

*1) $ABA$ is symmetric if $(ABA)^T = ABA$*

\begin{equation}
(ABA)^T = A^TB^TA^T = ABA
\end{equation}
and 
\begin{equation}
(BAB)^T = B^TA^TB^T = BAB
\end{equation}
since A and B are symmetric; $ A = A^T, B = B^T $. 

*2) $ABA$ is positive-definite if $\bold{x}^T (ABA)\bold{x} > 0$ for all $\bold{x} \in \mathbb{R}^n \backslash \{0\}$*

We define 
\begin{equation}
\bold{y}  = A\bold{x} 
\end{equation}
where $x\neq 0$. Since A is SPD, A is invertible, $y=Ax \neq 0$.
\begin{equation}
\bold{x}^T (ABA)\bold{x} = \bold{x}^T ABA\bold{x} = (A\bold{x})^T B (A\bold{x}) = \bold{y} ^TB\bold{y} 
\end{equation}
Since B is SPD, we have that 
$\begin{equation}
\bold{y} ^TB\bold{y}  > 0
\end{equation}$
for any  $\bold{y} \in \mathbb{R}^n \backslash \{0\}$.
Therefore
$\bold{x}^T (ABA)\bold{x} = \bold{y} ^TB\bold{y}  > 0$

The same result will hold for $BAB$. We define 
\begin{equation}
\bold{z}  = B\bold{x} 
\end{equation}
where $x\neq 0$. Since B is SPD, B is invertible, $Z=Bx \neq 0$.
\begin{equation}
\bold{x}^T (BAB)\bold{x} = \bold{x}^T BAB\bold{x} = (B\bold{x})^T A (B\bold{x}) = \bold{z} ^TA\bold{z} 
\end{equation}
Since A is SPD, we have that 
\begin{equation}
\bold{z} ^TA\bold{z}  > 0
\end{equation}
for any  $\bold{z} \in \mathbb{R}^n \backslash {0}$. Therefore
\begin{equation}
\bold{x}^T (BAB)\bold{x} = \bold{z} ^TA\bold{z}  > 0.
\end{equation}

$AB$ is, however, not a symmetric positive-definite matrix. Symmetry would require $(AB)^T=BA$. We have
\begin{equation}
(AB)^T = B^TA^T = BA
\end{equation}
Thus, we need
\begin{equation}
AB=BA
\end{equation}
It is known that matrices generally do not commute, meaning that $AB \neq BA$. Therefore $AB$ is not a $SPD$ matrix in genereal.

$J_{n}$ is a symmetric tridiagonal matrix, and the determinant of $J_{n}$ satisifes the following recurrence relation:
\begin{equation}
det(J_{n}-xI) = (x- \alpha_{n})det(J_{n-1} - xI) - \beta_{n}det(J_{n} - xI)

We can see that this is the same recurrance relation as the one given for $p_{n}(x)$
\begin{equation}
p_{n}(x) = (x-\alpha_{n})p_{n-1}(x) + \beta_{n}p_{n-2}(x)
\end{equation}

The characteristic polynomial of $J_{n}$ satisifies the same recurrence relation as $p_{n}(x)$ it follows that
\begin{equation}
det(J_{n}-xI) = p_{n}(x)
\end{equation}

The eigenvalues of the matrix $J_n$ is given by solving $det(J_{n}-xI) = 0$, the zeros of $p_{n}(x)$ is given by solving $P_{n}(x) = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Makes the tridiagonal matrix
def tridiagonal_matrix(n, alpha, beta):
    J = np.zeros((n,n))
    for i in range(n):
        J[i,i] = alpha[i]
        if i < n-1:
            J[i,i+1] = np.sqrt(beta[i+1])
            J[i+1,i] = np.sqrt(beta[i+1])
    return J


# Example values
n = 5
alpha = np.array([1,2,3,4,5])
beta = np.array([1,2,3,4,5,6])

# Constructs tridiagonal matrix
J_n = tridiagonal_matrix(n,alpha,beta)

# Finds eigenvalues
eigenvalues = np.linalg.eigvals(J_n)

# Plots the zeros (same as eigenvalues of J_n)
plt.grid()
plt.scatter(eigenvalues, np.zeros_like(eigenvalues))
plt.title(f'Zeros (eigenvalues) of the polynomial $p_n(x)$ for n={n}')
plt.xlabel('Eigenvalue (zero)')

plt.show()
```
